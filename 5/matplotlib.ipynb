```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
import pandas as pd
```

```python
#plotando dois graficos em conjunto e manipulando os eixos
x = np.arange(-5, 5, 0.001)
axes = plt.axes()
axes.set_xlim([-10,10])
axes.set_ylim([0, 1])
val_x = np.arange(-10, 11, 1)
val_y = np.arange(0, 1.1, 0.1)
axes.set_xticks(val_x)
axes.set_yticks(val_y)
axes.grid()
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1, 0.5))
plt.show()
```

```python
#plotando um grafico de setores
valores = [3, 5, 10, 1, 8]
cores = ['r', 'g', 'b', 'y', 'm']
labels = ['Kwid', 'Montana', 'Uno', 'Civic', 'Corolla']
plt.pie(valores, colors=cores, labels=labels)
plt.title('Carros no estacionamento')
plt.show()
```

```python
#plotando um grfico de barras
plt.bar(range(1,6), valores, color=cores)
plt.show()

```

```python
#plotando um grfico de dispersao
x = np.random.randn(200)
y = np.random.randn(200)

plt.scatter(x,y)
plt.show()
```

```python
#plotando um diagrama de caixa
dados = np.random.randint(0, 50000, 10000)
plt.boxplot(dados)
```

```python
import seaborn as sns
sns.set()
```

```python
#criando um dataframe
col = 'Civic Corolla Celta Palio'.split()

df = pd.DataFrame(col, index='A B C D'.split(), columns='Carros'.split())
consumoCidade = [12, 11, 8, 5,]
consumoEstrada = [18, 15, 10, 7]
ano = [2020, 2018, 2013, 2008]

df['ConsumoCidade'] = consumoCidade
df['ConsumoEstrada'] = consumoEstrada
df['Ano'] = ano
df
```

```python
#plotando o grafico
sns.distplot(df['ConsumoCidade'])
```

```python
sns.pairplot(df, hue='ConsumoCidade', height=2.5)
```
