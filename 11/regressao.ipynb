# Regress√£o Linear

```python
import numpy as np
import matplotlib.pyplot as plt

#criando uma fun√ß√£o linear
pesoVeiculo = np.random.normal(1500, 200, 1000)
consumoBase = 15 
desvioConsumo = np.random.normal(0, 1, 1000)

consumoVeiculo = consumoBase - (pesoVeiculo - 1200) * 0.01 + desvioConsumo

plt.scatter(pesoVeiculo, consumoVeiculo)
```

```python
from scipy import stats

#armazena os dados da funcao de regressao linear
slope, intercept, r_value, p_value, std_err = stats.linregress(pesoVeiculo, consumoVeiculo)
```

```python
#foi obtido um valor de R¬≤ de 0.81, o que √© satisfat√≥rio
r_value ** 2
```

```python
#criando a reta de regress√£o no gr√°fico
def predict(x):
    return slope * x + intercept

fitline = predict(pesoVeiculo)

plt.scatter(pesoVeiculo, consumoVeiculo)
plt.plot(pesoVeiculo, fitline, c='r')
plt.show()
```

# Regress√£o Polinomial

```python
np.random.seed(2)

#criando uma fun√ß√£o polinomial
pesoVeiculo = np.random.normal(1500, 300, 1000)
consumoBase = 15 
desvioConsumo = np.random.normal(0, 1, 1000)

consumoVeiculo = (consumoBase+desvioConsumo)*1000/pesoVeiculo

plt.scatter(pesoVeiculo, consumoVeiculo)
```

```python
#criando o polin√¥mio de regress√£o
x = np.array(pesoVeiculo)
y = np.array(consumoVeiculo)

p = np.poly1d(np.polyfit(x, y, 4))
```

```python
#plotando o polin√¥mio
xp = np.linspace(0, 3000, 100)
plt.scatter(x,y)
plt.plot(xp, p(xp), c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y, p(x))

print(r2)

#o r2 d√° um valor bem satisfat√≥rio (0.91)
```
