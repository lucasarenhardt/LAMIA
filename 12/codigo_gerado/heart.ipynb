```python
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
base = pd.read_csv('heart_failure_clinical_records_dataset.csv')
```

```python
#separando os previsores e as classes

previsores = base.iloc[:, 0:12].values
classe = base.iloc[:, 12].values

from sklearn.preprocessing import StandardScaler


#escalando os dados
scaler = StandardScaler()
previsores = scaler.fit_transform(previsores)
```

```python
#função para criar a rede neural

def criarede():
    classificador = Sequential()
    classificador.add(Dense(units=7, activation='relu', kernel_initializer='random_normal', input_dim=12))
    classificador.add(Dropout(0.2))
    classificador.add(Dense(units=7, activation='relu', kernel_initializer='random_normal'))
    classificador.add(Dropout(0.2))
    classificador.add(Dense(units=1, activation='sigmoid'))
    otimizador = keras.optimizers.Adam(learning_rate=0.001)
    classificador.compile(optimizer=otimizador, loss='binary_crossentropy', metrics=['binary_accuracy'])
    return classificador
```

```python
#treinando a rede neural com validação cruzada

classificador = KerasClassifier(model=criarede, epochs=100, batch_size=10)

resultados = cross_val_score(estimator=classificador,
                             X=previsores, y=classe,
                             cv=10, scoring='accuracy')
```

```python
#calculando a média da precisão e o desvio padrão

media = resultados.mean()
desvio = resultados.std()
print(media, desvio)
```
