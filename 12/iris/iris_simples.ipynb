# Iris simples

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
import keras.utils
import numpy as np
```

```python
#carregando o dataset
iris = load_iris()
previsores = iris.data
classe = iris.target
```

```python
#transfomando os dados para a classe dummy

classe_dummy = keras.utils.to_categorical(classe)
#iris setosa 1 0 0
#iris virginica 0 1 0
#iris versicolor 0 0 1
```

```python
#dividindo o conjunto de dados

X_train, X_test, y_train, y_test = train_test_split(previsores, classe_dummy, test_size=0.25)
```

```python
#criando as camadas da rede neural

classificador = Sequential()
classificador.add(Dense(units=4, activation='relu', input_dim=4))
classificador.add(Dense(units=4, activation='relu'))
classificador.add(Dense(units=3, activation='softmax'))
classificador.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
#realizando o treinamento

classificador.fit(X_train, y_train, batch_size=10, epochs=1000)
```

```python
#calculando a precisão

resultado = classificador.evaluate(X_test, y_test)
resultado
```

```python
previsoes = classificador.predict(X_test)
previsoes = (previsoes>0.5).astype(int)
```

```python
#transformando os dados para o índice em que está o '1'

y_test2 = [np.argmax(t) for t in y_test]
previsoes2 = [np.argmax(t) for t in previsoes]
```

```python
#imprimindo a matriz de confusão

from sklearn.metrics import confusion_matrix

matriz = confusion_matrix(previsoes2, y_test2)
matriz
```
